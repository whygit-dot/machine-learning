```python
!pip install transformers
import gradio as gr
from transformers import pipeline

# Load pre-trained summarization model from Hugging Face
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Summarization function
def summarize_article(article):
    summary = summarizer(article, max_length=200, min_length=50, do_sample=False)
    return summary[0]['summary_text']

# Create the Gradio interface
interface = gr.Interface(
    fn=summarize_article,
    inputs=gr.Textbox(label="Input Article", lines=10, placeholder="Paste your article here..."),
    outputs=gr.Textbox(label="Summary"),
    title="📰 News Summarization Tool",
    description="Enter a news article, and this tool will generate a concise summary of the article."
)

# Launch the interface
interface.launch()
```
